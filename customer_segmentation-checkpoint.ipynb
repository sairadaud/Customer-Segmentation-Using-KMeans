```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

X.head()
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8,5))

plt.plot(range(1,11), wcss, marker='o')

plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")

plt.savefig("elbow_method.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)

df['Cluster'] = kmeans.fit_predict(X_scaled)

df.head()
```

```python
df['Cluster'].value_counts().sort_index()
```

```python
plt.figure(figsize=(10,7))

sns.scatterplot(
    data=df,
    x='Annual Income (k$)',
    y='Spending Score (1-100)',
    hue='Cluster',
    palette='Set1',
    s=100
)

plt.title("Customer Segments using K-Means")

plt.savefig("customer_clusters.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
centers = scaler.inverse_transform(kmeans.cluster_centers_)

centers
```

```python
plt.figure(figsize=(10,7))

sns.scatterplot(
    data=df,
    x='Annual Income (k$)',
    y='Spending Score (1-100)',
    hue='Cluster',
    palette='Set1',
    s=100
)

plt.scatter(
    centers[:,0],
    centers[:,1],
    c='black',
    s=250,
    marker='X',
    label='Centroids'
)

plt.title("Customer Segmentation with Centroids")

plt.legend()

plt.savefig("cluster_centroids.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

X_pca = pca.fit_transform(X_scaled)
```

```python
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])

pca_df['Cluster'] = df['Cluster']

pca_df.head()
```

```python
plt.figure(figsize=(10,7))

sns.scatterplot(
    data=pca_df,
    x='PC1',
    y='PC2',
    hue='Cluster',
    palette='Set1',
    s=100
)

plt.title("Customer Segmentation using PCA")

plt.savefig("pca_clusters.png", dpi=300, bbox_inches='tight')

plt.show()
```
